```python
import pandas as pd, numpy as np

# create random dates
rng = pd.date_range('2024-01-01', periods=180, freq='D')

# fix random seed for reproducibility
np.random.seed(42)

# number of rows
rows = 2000

# create dataframe
df = pd.DataFrame({
    'order_id': np.arange(1, rows+1),
    'order_date': np.random.choice(rng, rows),
    'customer_id': np.random.randint(1000, 1200, rows),
    'product': np.random.choice(['Widget A','Widget B','Gadget C','Gizmo D'], rows),
    'category': np.random.choice(['Widgets','Gadgets','Accessories'], rows),
    'quantity': np.random.randint(1,6, rows),
    'price': np.random.uniform(5,100, rows).round(2),
    'discount': np.random.choice([0, 0.05, 0.1, 0.2], rows)
})

# add sales column
df['sales'] = (df['quantity']*df['price']*(1-df['discount'])).round(2)

# save to csv (optional)
df.to_csv('orders.csv', index=False)

# show shape
len(df)
```

```python
df[str_cols] = df[str_cols].apply(lambda s: s.str.strip())
df.head(3)
```

```python
df = df.drop_duplicates()
df.shape  # check new number of rows
```

```python
# check types of all columns
df.dtypes
```

```python
# fill any missing numeric values with 0 (just in case)
num_cols = ['quantity', 'price', 'discount', 'sales']
df[num_cols] = df[num_cols].fillna(0)

# recalculate sales just to be safe
df['sales'] = (df['quantity'] * df['price'] * (1 - df['discount'])).round(2)

df.head(3)
```

```python
# group sales by day
daily = df.groupby('order_date')['sales'].sum().reset_index()
daily.head(5)
```

```python
# group sales by category
cat_sales = df.groupby('category', as_index=False)['sales'].sum().sort_values('sales', ascending=False)
cat_sales
```

```python
# group sales by product
top_products = df.groupby('product', as_index=False)['sales'].sum().sort_values('sales', ascending=False).head(10)
top_products
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
plt.plot(daily['order_date'], daily['sales'], color='blue')
plt.title('Daily Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.xticks(rotation=25)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.bar(cat_sales['category'], cat_sales['sales'], color='green')
plt.title('Top Categories by Sales')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.bar(top_products['product'], top_products['sales'], color='orange')
plt.title('Top 10 Products by Sales')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.xticks(rotation=25)
plt.tight_layout()
plt.show()
```
